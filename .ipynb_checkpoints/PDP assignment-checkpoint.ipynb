```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_F(x):
    f = [0]*5
    f[0] = x[0]+x[1]+x[2]-13.5
    f[1] = 2*x[3]+4*x[0]+2*x[4]-227
    f[2] = x[4]+x[1]+2*x[2]-86.5
    f[3] = x[2]*x[3]/(x[1]*x[4])-0.929
    f[4] = ((12.2/100)**2)*x[1]*(x[4]**3)/(x[0]*x[4])-126.8

    return f
```

```python
def get_Z(x):
    z = [[1, 1, 1, 0, 0],
         [4, 0, 0, 2, 2],
         [0, 1, 2, 0, 1],
         [0, -(x[2]*x[3]/(x[1]**2*x[4])), x[3]/(x[1]*x[4]), x[2]/(x[1]*x[4]), -(x[2]*x[3]/(x[1]*x[4]**2))],
         [-(x[1]*x[3]**3/(x[0]*x[4]**2)*0.014884), x[3]**3/(x[0]*x[4])*0.014884, 0, 3*0.014884*x[1]*x[3]**2/(x[0]*x[4]), -(x[1]*x[3]**3/(x[0]*x[4]**2)*0.014884)]]
    
    return z
```

```python
x = [29]*5
f = get_F(x)
z = get_Z(x)
c = []
iter = []

for i in range(100):
    
    z_inv = np.linalg.inv(z)
    iter.append(i)
    
#     print("iter ", i+1, "\n\n")
#     print("X\n", x, "\n\n")
#     print("Z\n", z, "\n\n")
#     print("Z_inv\n", z_inv, "\n\n")
#     print("F\n", f, "\n\n")
#     print(" ========================================================\n\n")
    
    x_ = np.matmul(z_inv, ( np.negative(f) + np.matmul(z, x)))
    x = x_
    f = get_F(x_)
    z = get_Z(x_)
    c.append(x_)
    
print(len(iter), len(c[0]))
plt.figure(figsize=(20,10))
plt.plot(iter, [i[0] for i in c], [i[1] for i in c])
plt.show()
```
